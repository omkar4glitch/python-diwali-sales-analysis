**from the above graph we can see that most of the buyer are female and even the purchasing power of female is more than men.**

**###the amount is in the 10 to the power 7 as it is given on top of graph(i.e-Le7)**

**STATES*

```python
df.columns
```

```python
#total amount/sales  from top 10 states

states_sale=df.groupby(['State'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y="Amount",data=states_sale)
```

```python
#total numbers of orders from top 10 states

states_sales=df.groupby(['State'],as_index=False)["Orders"].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y="Orders",data=states_sales)
```

**from the above graphs we can see that the most of the orders and total sale/amt are from up,maharastra karnataka respectively** 

**marital status**

```python
ax=sns.countplot(x="Marital_Status",data=df)
sns.set(rc={"figure.figsize":(7,5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
state_sales=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(7,5)})
sns.barplot(x="Marital_Status",y="Amount",data=state_sales,hue="Gender")
```

**from the above graph we can see that the most of the buyers are married(women) and have high purchasing power.**

**OCCUPATION*

```python
df.columns
```

```python
ax=sns.countplot(x="Occupation",data=df)
sns.set(rc={'figure.figsize':(20,5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
occupation_sales=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Occupation",y="Amount",data=occupation_sales)
```

**from the abouve graphs we can see that the most of the buyer  are working in it, healthcare and aviation sector.**

**PRODUCT CATEGORY*

```python
ax=sns.countplot(x="Product_Category",data=df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
products_sales=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending =False).head(10)
sns.barplot(x="Product_Category",y="Amount",data=products_sales)
```

**from the above graph we can see that the most sold products are from food,clothing and electronics category.**

```python
productid_orders=df.groupby(['Product_ID'],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending =False).head(10)
sns.barplot(x='Product_ID',y="Orders",data=productid_orders)
```

```python
#top 10 most sold products(same thing as above)
fig1,ax1=plt.subplots(figsize=(12,4))
df.groupby("Product_ID")["Orders"].sum().nlargest(10).sort_values(ascending =False).plot(kind='bar')
```

**conclusion:***

**married women age group 26-35 from up,maharastra and karnataka working in it,healthcare and aviation are most likely to buy products from food,clothing and electornics category.**
